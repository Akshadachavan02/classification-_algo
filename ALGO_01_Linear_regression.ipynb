### Step 2 :Find out the target variable
** In boston hosue prediction dataset the MEDV is a target varivable .i.e median of house price **

```python
#x= feachures y= target varible
x = boston_df.drop(["MEDV"],axis=1)
y = boston_df["MEDV"]
x.shape
```

```python
## output variable with one colunm
y.shape
```

### Split the dataset as in training and testing part  
##### we are using 80% fro training and 20% for testing 

```python
from sklearn.model_selection import train_test_split
```

###### ***As we have to pass the parameter to test size in btw of 1-0

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
print("shape of the train dataset feachure")
print(x_train.shape)
print("shape of the test dataset feachure")
print(x_test.shape)
```

```python
print("shape of the train dataset target")
print(y_train.shape)
print("shape of the test dataset target")
print(y_train.shape)
```

```python
from sklearn.linear_model  import LinearRegression
```

```python
#iNITIATING THE ESTIMATING OBJECT 
model_linreg = LinearRegression()
```

```python
model_linreg.fit(x_train,y_train) ##supervised machine learning approach
```

```python
print("The intercept for multiple linear regression on boston dataset is :: ",model_linreg.intercept_)
```

```python
print("the linear coefficient for multiple linear regression model are ::")
print(model_linreg.coef_)
```

```python
x.columns
```

```python
y_test_pred = model_linreg.predict(x_test)
```

```python
y_test_pred
```

```python
df = pd.DataFrame({"Achual":y_test , "predicted":y_test_pred})
```

#### Step 3:RMSE and R2 error

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
print("The RMSE  value of the given data set is ::")
print(np.sqrt(mean_squared_error(y_test,y_test_pred)))
```

```python
print(" The R2 error value of the dataset is::")
print(model_linreg.score(x_test,y_test))
```

###  From above value we can say that for 71% of the data mode is fit for it

### lets find out the model is overfiting underfitting or appropriate

```python
## pass the train value in order to calculate the y_train predicted
y_train_predict = model_linreg.predict(x_train)
```

```python
## calculate the RMSE value for the train datase
print("The RMSE  for the train dataet is ::")
print(np.sqrt(mean_squared_error(y_train,y_train_predict)))
```

##### from above we can that the there isno such big differance in train and test RMSE  value so model isappropriate

# regularization

```python
from sklearn.linear_model import Ridge
```

```python
ridge_mod = Ridge(alpha=1.5) #3estimating the object
```

```python
ridge_mod.fit(x_train,y_train)
```

```python
y_predict_ridge_test = ridge_mod.predict(x_test)
```

```python
y_predict_ridge_train = ridge_mod.predict(x_train)
```

```python
print("The RMSE  for the ridge regression on  train dataet is ::")
print(np.sqrt(mean_squared_error(y_train,y_predict_ridge_train)))
```

```python
print("The RMSE  for the ridge regression on  train dataet is ::")
print(np.sqrt(mean_squared_error(y_test,y_predict_ridge_test)))
```

### for lasso regulariation model

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 0.01)
```

```python
lasso_model.fit(x_train,y_train)
```

```python
lasso_y_predict_test = lasso_model.predict(x_test)
lasso_y_predict_train = lasso_model.predict(x_train)
```

```python
print("The RMSE  for the lasso regression on  train dataet is ::")   ##alpha=1 ==>5.10796815965904
print(np.sqrt(mean_squared_error(y_train,lasso_y_predict_train)))
```

```python
print("The RMSE  for the lasso regression on  test dataet is ::")    ##alpha=1 ==>5.832100125207783
print(np.sqrt(mean_squared_error(y_test,lasso_y_predict_test)))
```
