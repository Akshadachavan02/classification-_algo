```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Consumo_cerveja.csv")
```

```python
new_df.info()
```

```python
new_df['Temperatura Media (C)'] = new_df['Temperatura Media (C)'].str.replace(",",".")

new_df
```

```python
new_df['Temperatura Minima (C)'] = new_df['Temperatura Minima (C)'].str.replace(',', '.')
```

```python
new_df["Temperatura Maxima (C)"] = new_df["Temperatura Maxima (C)"].str.replace(",",".")
```

```python
new_df["Precipitacao (mm)"] = new_df["Precipitacao (mm)"].str.replace(",",".")
```

```python
new_df
```

```python
##let's change the colum name for naming convention 
new_df.rename(columns = {"Temperatura Media (C)": "Average_temperature",
               "Temperatura Minima (C)":"Minimum-temperature",
               "TemperaturaMaxima (C)": "Maximum_temperature",
               "Precipitacao (mm)": "Percipitation",
               "Final de Semana": "is_weekend",
               "Consumo de cerveja (litros)": "Beer_consumption"},inplace=True)
new_df
```

```python
## SET DATE COLUMN AS INDEX
##new_df.set_index("Data")
```

```python
## using the dlimeter we extracted the day and date from data colunm
new_df["Data"] = pd.to_datetime(new_df["Data"])
new_df["Day"] = new_df["Data"].dt.day_name()
new_df["Month"] = new_df["Data"].dt.month
```

```python
new_df
```

```python
## check the duplicate values in dataset 
duplicated_row = new_df[new_df.duplicated()]
print(duplicated_row)               
```

```python
new_df.info()
```

```python
new_df["Average_temperature"] = new_df["Average_temperature"].astype(float)
new_df["Minimum-temperature"] = new_df["Minimum-temperature"].astype(float)
new_df["Temperatura Maxima (C)"] = new_df["Temperatura Maxima (C)"].astype(float)
new_df["Percipitation"] = new_df["Percipitation"].astype(float)
```

```python
print(new_df.dtypes)
```

```python
##Analyze features with outlier values 
plt.figure(figsize=(10, 6))  # Increase figure size

# Rotate x-axis labels
plt.xticks(rotation=45)

# Decrease font size of x-axis labels
plt.xticks(fontsize=8)
sns.boxplot(data = new_df)

plt.show()
```

**we can see that the precipitation column has the so many outliers



```python
## let's find the  correlation between the variables 
## heatmap is the best way to find correlation
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.heatmap(new_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")  # 'annot=True' to display correlation values, 'fmt=".2f"' for 2 decimal places
plt.title('Correlation Heatmap')  # Set title
plt.show()
```

```python
#Split the dataset for training and testing 
from sklearn.model_selection import train_test_split
```

```python
# x is out frachres and y is our target 
x = new_df.drop(columns = ["Beer_consumption","Data","Day"])
y = new_df["Beer_consumption"]
print(x.shape)
print(y.shape)
```

```python
#now lets devide the dataset into the train and test split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Train a linear regression model and print the intercept and coefficients
from sklearn.linear_model import LinearRegression
```

```python
regg_model = LinearRegression()
```

```python
regg_model.fit(x_train,y_train)
```

```python
y_predict_test =regg_model.predict(x_test)
```

```python
y_predict_test
```

```python
df2 = pd.DataFrame({"Achual":y_test,"Predicted":y_predict_test})
df2
```

```python
# intercept for the model 
print('The intercept for the given model is ::',regg_model.intercept_)
```

```python
# coefficient for the model is 
print("coefficient of the model is :",regg_model.coef_)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
print("the resme value ofthe datset is ")
print(np.sqrt(mean_squared_error(y_test,y_predict_test)))
```

```python
## r2value is 
print(" The R2 error value of the dataset is::")
print(regg_model.score(x_test,y_test))
```
