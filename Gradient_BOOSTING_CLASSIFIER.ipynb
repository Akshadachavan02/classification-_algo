```python
from sklearn.datasets import load_iris
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
iris_bunch = load_iris()
iris_df = pd.DataFrame(iris_bunch.data,columns =iris_bunch.feature_names)
iris_df["class"] = iris_bunch.target
iris_df
```

```python
x= iris_df.drop(columns="class")
y = iris_df["class"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state =21)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbd_mod = GradientBoostingClassifier(n_estimators=20,learning_rate=0.3)
gbd_mod.fit(x_train,y_train)
```

```python
y_pred_test = gbd_mod.predict(x_test)
y_pred_train = gbd_mod.predict(x_train)
```

```python
print("accuracy score for the test data set is:")
print(accuracy_score(y_test,y_pred_test))
```

```python
print("accuracy score for the train data set is:")
print(accuracy_score(y_train,y_pred_train))
```

```python
print("classifier report  for the train data set is:")
print(classification_report(y_train,y_pred_train))
```

```python
print("classifier report  for the test data set is:")
print(classification_report(y_test,y_pred_test))
```
